```python
# kirim data file
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
from google.colab import files
uploaded = files.upload()
```

```python
# membaca file
df = pd.read_csv('stock.csv')
```

```python
# menampilkan file
df.head(13)
```

```python
# Histogram Purchase Price
sns.histplot(df['Purchase Price'],kde=True)
plt.xlabel('Purchase Price', fontsize = 10)
plt.ylabel('Frequesi', fontsize = 10)
plt.title('Data Histogram Purchase Price', fontsize = 15)
plt.show()
```

```python
# Histogram Current Price
sns.histplot(df['Current Price'],kde=True)
plt.xlabel('Current Price', fontsize = 10)
plt.ylabel('Frequesi', fontsize = 10)
plt.title('Data Histogram Current Price', fontsize = 15)
plt.show()
```

```python
# Histogram Gain/Loss
sns.histplot(df['Gain/Loss'],kde=True)
plt.xlabel('Gain/Loss', fontsize = 10)
plt.ylabel('Frequesi', fontsize = 10)
plt.title('Data Histogram Gain/Loss', fontsize = 15)
plt.show()
```

```python
# Dividend/share top 5
sortedtop5_df = df.sort_values('Dividend/Share', ascending=False)
top_5_dividends = sortedtop5_df.head(5)
print(top_5_dividends[['Stock Name', 'Symbol']])
```

```python
# Market Value top 5
sortedtop5_df = df.sort_values('Market Value', ascending=False)
top_5_dividends = sortedtop5_df.head(5)
print(top_5_dividends[['Stock Name', 'Symbol']])
```

```python
# Annual yield top 7
sortedtop7_df = df.sort_values('Annual Yield', ascending=False)
top_5_dividends = sortedtop7_df.head(7)
print(top_5_dividends[['Stock Name', 'Symbol']])
```
